Estadísticas de los jugadores/as e implementación de la función que devuelve una estructura de datos asociada a cada jugador o jugadora.

```python
from hockey import Estructura
names = """Agustin, Yanina, Andrés, Ariadna, Bautista, CAROLINA,
CESAR, David, Diego, Dolores, DYLAN, ELIANA, Emanuel, Fabián, Noelia,
Francsica', FEDERICO, Fernanda, GONZALO, Nancy """
goals = [0, 10, 4, 0, 5, 14, 0, 0, 7, 2, 1, 1, 1, 5, 6, 1, 1, 2, 0,11]
goals_avoided = [0, 2, 0, 0, 5, 2, 0, 0, 1, 2, 0, 5, 5, 0, 1, 0, 2,3, 0, 0]
assists = [0, 5, 1, 0, 5, 2, 0, 0, 1, 2, 1, 5, 5, 0, 1, 0, 2, 3, 1,0]

players=Estructura.generar_estructura(names,goals,goals_avoided,assists)
print(f'Jugadores/as del equipo: {players}')
```

Función que devuelve el nombre del jugador/a con más goles del equipo.

```python
from hockey import premios
goleador,max_goles=premios.obtener_goleador(players)
print(f'El/La goleador/a es: {goleador} con una cantidad de {max_goles} goles.')
```

Función que devuelve el nombre del jugador/a más influyente según el promedio ponderado

```python
influyente=premios.obtener_influyente(players)
print(f'El/La jugador/a mas influyente es {influyente}')
```

Función que devuelve el promedio general de goles del equipo en la temporada (25 partidos).

```python
promedio_general=premios.obtener_promedio_general(players)
print(f'El promedio general del equipo es: {promedio_general} goles por partido')
```

Función que devuelve el promedio individual de goles del goleador/a en la temporada (25 partidos).

```python
promedio_goleador=premios.obtener_promedio_individual(players)
print(f'El promedio individual del goleador/a es de: {promedio_goleador} goles por partido')
```
